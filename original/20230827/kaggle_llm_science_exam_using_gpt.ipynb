```python
# 更新版
# {"prompt": "Q: <prompt>, A: <answer>, B: <answer>, C: <answer>, D: <answer>, E: <answer>", "completion": "<answer>"}
preprocessedData = list()
for _, i in rawData.iterrows():
    preprocessedData.append(
        {
            "prompt": """Q: {}
            A: {}
            B: {}
            C: {}
            D: {}
            E: {}""".format(i["prompt"], i["A"], i["B"], i["C"], i["D"], i["E"]),
            "completion": i["answer"]
        }
    )
```

```python
# 定義成表格
preprocessedData = pd.DataFrame(preprocessedData)
```

```python
# 切割成訓練、驗證與測試資料
trainData, testData = train_test_split(preprocessedData, test_size = 0.2, stratify = preprocessedData["completion"])
```

```python
trainData.to_json("newTrainData.jsonl", orient = "records", lines = True)
```

```python
# 轉換成 chat message 格式
# {"messages": [{"role": "system", "content": "Marv is a factual chatbot that is also sarcastic."}, {"role": "user", "content": "What's the capital of France?"}, {"role": "assistant", "content": "Paris, as if everyone doesn't know that already."}]}
chatMessage = list()
for _, i in preprocessedData.iterrows():
    chatMessage.append(
        {"messages": [
            {"role": "user", "content": i["prompt"]},
            {"role": "assistant", "content": i["completion"]}
        ]}
    )
```

```python
# 將資料彙整成 JSONL File
import json
with open("chatTrainData.jsonl", "w") as f:
    for entry in chatMessage:
        json.dump(entry, f)
        f.write("\n")
```

```python
openai.File.create(open("chatTrainData.jsonl", "rb"), purpose = "fine-tune")
```
